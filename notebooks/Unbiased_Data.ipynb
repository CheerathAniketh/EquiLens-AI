```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n = 1000

gender = np.random.choice([0,1], size=n)
age = np.random.randint(21,50,size=n)
experience = np.random.randint(0,20,size=n)
education = np.random.choice([0,1,2], size=n)
```

```python
# IMPORTANT: hiring independent of gender
hired = (experience + education + np.random.randn(n)*2 > 10).astype(int)

df = pd.DataFrame({
    "gender": gender,
    "age": age,
    "experience": experience,
    "education": education,
    "hired": hired
})
```

```python
df.to_csv("true_unbiased.csv", index=False)
```
